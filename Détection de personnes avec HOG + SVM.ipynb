Exemple de détection avec HOG + SVM

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

Lecture des images en entrée. (Images provenant PETS 2006, http://www.cvg.reading.ac.uk/PETS2006/data.html)

```python
image1 = (cv2.imread('in000235.jpg',cv2.IMREAD_GRAYSCALE))
image2 = (cv2.imread('in000216.jpg',cv2.IMREAD_GRAYSCALE))
```

Initialisation du détecteur.

```python
hog = cv2.HOGDescriptor() #64 x 128 pixels
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
```

Détection multi-échelles et affichage.

```python
(rects, weights) = hog.detectMultiScale(image1, winStride = (4,4), padding = (8,8), scale = 1.05)

#Affichage
for (x, y, w, h) in rects:
    cv2.rectangle(image1, (x, y), (x + w, y + h), (255, 255, 255), 2)
    
plt.figure(figsize = (8,8))
plt.imshow(image1, cmap = plt.get_cmap('gray'))
plt.show()
```

```python
print(weights) #Score des détections
```

Détection multi-échelles et affichage.

```python
(rects, weights) = hog.detectMultiScale(image2, winStride=(2,2), padding= (8,8), scale=1.05)

#Affichage
for (x, y, w, h) in rects:
    cv2.rectangle(image2, (x, y), (x + w, y + h), (255, 255, 255), 2)
    
plt.figure(figsize = (8,8))
plt.imshow(image2, cmap = plt.get_cmap('gray'))
plt.show()
```

```python
print(weights) #Score des détections
```
